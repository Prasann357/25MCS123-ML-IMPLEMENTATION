```python
#Neuron with sigmoid activation
import numpy as np
import matplotlib.pyplot as plt

# Neuron with sigmoid activation
def neuron_sigmoid(x):
    w = 2
    b = 1
    z = w * x + b
    return 1 / (1 + np.exp(-z))

x = np.linspace(-5, 5, 200)
y = neuron_sigmoid(x)

plt.plot(x, y, label="Sigmoid Neuron", linewidth=2)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Neuron with Sigmoid Activation")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Neuron without Sigmoid Activation
import numpy as np
import matplotlib.pyplot as plt

# Neuron without activation (pure linear)
def neuron_linear(x):
    w = 2
    b = 1
    return w * x + b

x = np.linspace(-5, 5, 200)
y = neuron_linear(x)

plt.plot(x, y, label="Linear Neuron", linewidth=2)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Neuron without Sigmoid Activation (Linear Output)")
plt.legend()
plt.grid(True)
plt.show()
```
